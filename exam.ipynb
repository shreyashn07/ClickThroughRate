```
4. Identify the model:i have went with linear regression
```

```
data=df.select(['offerid','countrycode','category','merchant','utcdate','rating'])
```

```
from pyspark.sql.functions import unix_timestamp
```

```
data = data.withColumn('unix', unix_timestamp('utcdate'))
```

```
data.show(1)
```

```
train_data,test_data=data.randomSplit([0.7,0.3])
```

```
Building the Model
```

```
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import VectorAssembler,StringIndexer,OneHotEncoderEstimator
from pyspark.ml import Pipeline
```

```
country_code_indexer = StringIndexer(inputCol='countrycode',outputCol='country_code_index',handleInvalid='keep')
merchant_indexer=StringIndexer(inputCol='merchant',outputCol='merchant_index',handleInvalid='keep')
```

```
data_encoder = OneHotEncoderEstimator(inputCols=['merchant_index'],
                                      outputCols=['merchant_vec'],
                                      handleInvalid='keep')
```

```
I chose merchant,country,category and time,model was stuck and unable product results hence i reduced the variables to merchant and category
The logic behind choosing this variables offers are attractive in certain categories and it becomes the major turning point for customer to click or not to click
```

```

assembler = VectorAssembler(inputCols=["merchant_vec","category"],
                            outputCol="features")
```

```
lr_model = LogisticRegression(labelCol='rating')
```

```
pipe = Pipeline(stages=[merchant_indexer,data_encoder,assembler,lr_model])
```

```
fit_model=pipe.fit(train_data)
```

```
# Store the results in a dataframe

results = fit_model.transform(test_data)
```

```
results.groupby('prediction').count().show()
```

```
Evaluating the model
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
AUC_evaluator = BinaryClassificationEvaluator(rawPredictionCol='prediction',labelCol='rating',metricName='areaUnderROC')
```

```
AUC = AUC_evaluator.evaluate(results)
```

```
print("The area under the curve is {}".format(AUC))
```

```
Confusion Matrix
```

```
from sklearn.metrics import confusion_matrix
```

```
y_true = results.select("rating")
y_true = y_true.toPandas()

y_pred = results.select("prediction")
y_pred = y_pred.toPandas()

cnf_matrix = confusion_matrix(y_true, y_pred)
print("Below is the confusion matrix \n {}".format(cnf_matrix))
```
